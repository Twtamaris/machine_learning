```python
# this is fireworks
import cv2
import mediapipe as mp
import random

mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands

# Initialize OpenCV and Mediapipe
cap = cv2.VideoCapture(0)
hands = mp_hands.Hands()

# Set up the screen
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Firework properties
firework_radius = 5
firework_colors = [(255, 0, 0), (0, 255, 0), (0, 0, 255), (255, 255, 0), (255, 0, 255), (0, 255, 255)]
fireworks = []

while True:
    success, image = cap.read()

    # Flip the image horizontally for natural hand movements
    image = cv2.flip(image, 1)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Process hand landmarks
    results = hands.process(image_rgb)

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            # Get index finger tip coordinates
            index_finger_x = int(hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].x * width)
            index_finger_y = int(hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].y * height)

            # Check if finger is up (when the y-coordinate is below a threshold)
            # if index_finger_y < height // 2:
                # Create a new firework at the finger position
            firework = {
                "position": [index_finger_x, index_finger_y],
                "color": random.choice(firework_colors),
                "life": random.randint(50, 100)
            }
            fireworks.append(firework)

    # Update and draw fireworks
    for firework in fireworks:
        firework["life"] -= 1

        if firework["life"] <= 0:
            fireworks.remove(firework)
        else:
            # Draw firework
            cv2.circle(image, (firework["position"][0], firework["position"][1]), firework_radius, firework["color"], -1)

    # Display the image
    cv2.imshow('Fireworks Art', image)

    # Exit the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture and destroy windows
cap.release()
cv2.destroyAllWindows()
```

```python
import pygame
import mediapipe as mp
import cv2
import random

# Constants for the window size
WINDOW_WIDTH = 800
WINDOW_HEIGHT = 600

# Initialize Pygame
pygame.init()
window = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
clock = pygame.time.Clock()

# Initialize Mediapipe for hand tracking
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=False, max_num_hands=2)  # Increase the maximum number of hands

# Colors for the fireworks
FIREWORKS_COLORS = [(255, 0, 0), (0, 255, 0), (0, 0, 255), (255, 255, 0), (255, 0, 255)]

# Fireworks structure
fireworks = []

capture = cv2.VideoCapture(0)

# Game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_ESCAPE:
                running = False

    # Get the current frame from the webcam
    ret, frame = capture.read()
    frame = cv2.flip(frame, 1)

    # Convert the frame to RGB and pass it to Mediapipe
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(frame_rgb)

    # Clear the screen
    window.fill((0, 0, 0))

    # Check if hands are detected
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            # Get the hand landmarks
            landmarks = hand_landmarks.landmark

            # Get the tips of all fingers
            finger_tips = []
            for finger_tip in mp_hands.HandLandmark:
                x = int(landmarks[finger_tip].x * WINDOW_WIDTH)
                y = int(landmarks[finger_tip].y * WINDOW_HEIGHT)
                finger_tips.append((x, y))

            # Create fireworks at each finger position
            for finger_tip in finger_tips:
                firework = {
                    "position": [finger_tip[0], finger_tip[1]],
                    "color": random.choice(FIREWORKS_COLORS),
                    "life": random.randint(50, 100)
                }
                fireworks.append(firework)

    # Draw the fireworks
    for firework in fireworks:
        # Randomly choose a color for each firework
        color = firework["color"]

        # Draw the circle
        pygame.draw.circle(window, color, firework["position"], 3)

        # Move the firework upwards
        firework["position"][1] -= 5

        # Decrease the life of the firework
        firework["life"] -= 1

        # Remove the firework if its life reaches 0
        if firework["life"] <= 0:
            fireworks.remove(firework)

    pygame.display.flip()
    clock.tick(30)

# Clean up
capture.release()
pygame.quit()
```
