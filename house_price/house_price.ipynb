```python
null_percent = dataset.isnull().sum()/dataset.shape[0]*100
# missing value percentage
```

```python
column_to_drop = dataset.isnull().sum()[null_percent>50].keys()
```

```python
dataset = dataset.drop(columns=column_to_drop)
```

```python
dataset.shape
```

```python
dataset.columns[dataset.isnull().any()]
```

## Add columns mean to numerical columns with missing values

```python
# columns with numerical values
num_dataset = dataset.select_dtypes(include=['int64','float64'])
```

```python
# columns with missing values
num_dataset.columns[num_dataset.isnull().any()]
```

```python
dataset['LotFrontage'].fillna(dataset['LotFrontage'].mean(),inplace=True)
dataset['MasVnrArea'].fillna(dataset['MasVnrArea'].mean(),inplace=True)
dataset['GarageYrBlt'].fillna(dataset['GarageYrBlt'].mean(),inplace=True)
```

## Add columns to categorical columns

```python
object_column = dataset.select_dtypes(include=['object'])
```

```python
dataset['MasVnrType'].mode()[0]
```

```python
dataset['MasVnrType'].fillna(dataset['MasVnrType'].mode()[0],inplace=True)
dataset.MasVnrType
```

```python
list2 = ['MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1',
       'BsmtFinType2', 'Electrical', 'FireplaceQu', 'GarageType',
       'GarageFinish', 'GarageQual', 'GarageCond']
```
