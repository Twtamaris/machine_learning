# Image Processing
1. Resize the images
2. Convert the images to numpy array

```python
Image.open('with_mask/with_mask_1.jpg').resize((100, 100))
```

```python
with_mask
```

```python
data = []
for i in with_mask:
    img = Image.open('with_mask/'+i)
    img = img.resize((32, 32))
    img = img.convert('RGB')
    # Here we have used np.asarray() rather than np.array()
    # because np.array() will create a copy of the object
    # whereas np.asarray() will create a reference to the object
    # which will be more efficient and memory saving

    img = np.asarray(img)
    data.append(img)
    
```

```python
for i in without_mask:
    img = Image.open('without_mask/'+i)
    img = img.resize((32, 32))
    img = img.convert('RGB')
    img = np.asarray(img)
    data.append(img)
```

Building the Convolutional Neural Network(CNN)

```python
import tensorflow as tf
from tensorflow import keras
```

```python
# one for mask and one for without mask
num_of_classes = 2
# Define an object
model = keras.Sequential()
# Adding the first CNN layer
model.add(keras.layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)))
# Adding the second layer with max pooling
model.add(keras.layers.MaxPooling2D((2, 2)))
# Adding the third CNN layer
model.add(keras.layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
# Adding the fourth layer with max pooling
model.add(keras.layers.MaxPooling2D((2, 2)))

# Adding the flatten layer
model.add(keras.layers.Flatten())
# Adding the first dense layer
# dense layer means all the neurons of this layers is 
# connected to all the neurons of the previous layer
model.add(keras.layers.Dense(128, activation='relu'))
# Adding Droupout layer
# Droupout layer is used to prevent overfitting
model.add(keras.layers.Dropout(0.5))

model.add(keras.layers.Dense(64, activation='relu'))
model.add(keras.layers.Dropout(0.5))

# Adding the second dense layer(output layer)
model.add(keras.layers.Dense(num_of_classes, activation='sigmoid'))

```
