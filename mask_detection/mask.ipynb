```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

```

```python
# import cv2_show
from PIL import Image
# import train_test_split
from sklearn.model_selection import train_test_split


```

```python
with_mask = os.listdir('with_mask')
print(len(with_mask))
```

```python
without_mask = os.listdir('without_mask')
print(len(without_mask))
```

```python
with_mask, without_mask
```

```python
# create the labels
with_mask_labels = np.zeros(len(with_mask))
without_mask_labels = np.ones(len(without_mask))
```

```python
with_mask_labels, without_mask_labels
```

```python
labels = np.concatenate((with_mask_labels, without_mask_labels), axis=0)
```

```python
# displaying the mask image
img = mpimg.imread('with_mask/with_mask_1.jpg') 
plt.imshow(img)
plt.show()
```

```python
img = mpimg.imread('without_mask/without_mask_1.jpg')
plt.imshow(img)
plt.show()
```

# Image Processing
1. Resize the images
2. Convert the images to numpy array

```python
Image.open('with_mask/with_mask_1.jpg').resize((100, 100))
```

```python
with_mask
```

```python
data = []
for i in with_mask:
    img = Image.open('with_mask/'+i)
    img = img.resize((32, 32))
    img = img.convert('RGB')
    # Here we have used np.asarray() rather than np.array()
    # because np.array() will create a copy of the object
    # whereas np.asarray() will create a reference to the object
    # which will be more efficient and memory saving

    img = np.asarray(img)
    data.append(img)
    
```

```python
for i in without_mask:
    img = Image.open('without_mask/'+i)
    img = img.resize((32, 32))
    img = img.convert('RGB')
    img = np.asarray(img)
    data.append(img)
```

```python
data
```

```python
len(data), type(data), data[0].shape
# here 128, 128 is the size of the image and 3 is the number of channels
```

# Converting the images list to numpy arrrays


```python
labels
# labels 1 mean that the person is wearing mask
```

```python
array_data = np.array(data)
```

```python
array_data = array_data/255.0
```

```python
array_data
```

```python
type(array_data), array_data.shape, labels.shape, type(labels)
```

```python
# split the data into train and test
x_train, x_test, y_train, y_test = train_test_split(array_data, labels, test_size=0.2, random_state=42)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

Building the Convolutional Neural Network(CNN)

```python
import tensorflow as tf
from tensorflow import keras
```

```python
# one for mask and one for without mask
num_of_classes = 2
# Define an object
model = keras.Sequential()
# Adding the first CNN layer
model.add(keras.layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)))
# Adding the second layer with max pooling
model.add(keras.layers.MaxPooling2D((2, 2)))
# Adding the third CNN layer
model.add(keras.layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
# Adding the fourth layer with max pooling
model.add(keras.layers.MaxPooling2D((2, 2)))

# Adding the flatten layer
model.add(keras.layers.Flatten())
# Adding the first dense layer
# dense layer means all the neurons of this layers is 
# connected to all the neurons of the previous layer
model.add(keras.layers.Dense(128, activation='relu'))
# Adding Droupout layer
# Droupout layer is used to prevent overfitting
model.add(keras.layers.Dropout(0.5))

model.add(keras.layers.Dense(64, activation='relu'))
model.add(keras.layers.Dropout(0.5))

# Adding the second dense layer(output layer)
model.add(keras.layers.Dense(num_of_classes, activation='sigmoid'))

```

```python
model.summary()
```

# Training the Convolutional Neural Network(CNN)

```python
# compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train,epochs= 5, validation_split=0.1 )
```

# Model Evaluation and Prediction

```python
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=1)
```

```python
test_acc, test_loss
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred = np.argmax(y_pred, axis=1)
```

```python
y_pred
```

```python
y_test
```

```python
# import confusion_matrix
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
# plot the loss value
h = history
plt.plot(h.history['loss'], label='train loss')
plt.plot(h.history['val_loss'], label='validation loss')
plt.legend()
plt.show()

# plot the accuracy values
plt.plot(h.history['accuracy'], label='train accuracy')
plt.plot(h.history['val_accuracy'], label='validation accuracy')
plt.legend()
plt.show()
```
